```python
#Generating QR code of a fixed size 

qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_L,
    box_size=20,
    border=4,
)
qr.add_data('Sent to: <Name>') #17 characters
qr.make(fit=True)
img = qr.make_image()
img.save('imagename.jpg')
img.thumbnail((512,512))    #sizing it to 512x512
img_props(img, "qr") #prints size of the image
img
```

### Arnold Scrambling
Performing Arnolds Scrambling using Arnolds Cat Map

```python
qr_img = array(Image.open("img\watermark.jpg"))
N = qr_img.shape[0]

x,y = meshgrid(range(N), range(N))
#arnold scrambling
xmap = (2*x+y) % N
ymap = (x+y) % N

for i in range(N):
    result = Image.fromarray(qr_img)
    result.save("img/transformsWatermark/watermark_%03d.png" % i)
    #performing the scrambling
    qr_img = qr_img[xmap,ymap]  
```

Function to print properties of n-dimensional arrays

```python
def ndarray_props(img_nd, name):
    print("________")
    print("Properties of ", name)
    print("shape: ", img_nd.shape)
    print("size: ", img_nd.size)
    print("data type: ", img_nd.dtype)
    print("\n")

ndarray_props(watermark_img, "watermark_img")
ndarray_props(cover_img, "cover_img")
```
